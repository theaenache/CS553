```python
from transformers import pipeline

pipe = pipeline("image-classification", model="sharren/vit-dropout-v9")
```

```python
from datasets import load_dataset

ds = load_dataset("sharren/SkinCancerClassification")
```

```python
# prompt: use the pipeline and dataset above to do inference for first 3 examples

for i in range(3):
  image = ds['train'][i]['image']
  results = pipe(image)
  print(f"Image {i+1}:")
  for result in results:
    print(f"Label: {result['label']}, Score: {result['score']}")
  print("---")
```
